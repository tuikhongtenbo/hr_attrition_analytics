## 1.5. Outlier Detection & Treatment

### Outlier Detection

```python
numeric_cols.remove('Employee ID')
```

```python
df_outlier = df_no_id.copy()
outlier_summary = {}

print("=" * 70)
print("OUTLIER DETECTION (IQR METHOD)")
print("=" * 70)

for col in numeric_cols:
    data = df_outlier[col].dropna()

    # IQR METHOD
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers_iqr = ((data < lower) | (data > upper)).sum()

    # Save summary
    outlier_summary[col] = {
        "IQR_outliers": outliers_iqr,
        "IQR_lower": lower,
        "IQR_upper": upper
    }

    print(f"\n{col}")
    print("-" * 40)
    print(f"IQR method outliers     : {outliers_iqr}")
```

```python
outlier_cols = [
    col for col, info in outlier_summary.items()
    if (info["IQR_outliers"] > 0)
]

print(f"Columns with outliers: {len(outlier_cols)}")
print(outlier_cols)

if len(outlier_cols) == 0:
    print("No outliers detected!")
else:
    n_features = len(outlier_cols)
    n_rows = math.ceil(n_features / 2)

    fig = plt.figure(figsize=(14, 3.5 * n_rows))
    gs = GridSpec(n_rows, 2, figure=fig)

    row = col_pos = 0

    for i, col in enumerate(outlier_cols):
        data = df_outlier[col].dropna()
        lower = outlier_summary[col]["IQR_lower"]
        upper = outlier_summary[col]["IQR_upper"]

        # Nếu là cột cuối và số lượng lẻ -> căn giữa
        if (i == n_features - 1) and (n_features % 2 == 1):
            ax = fig.add_subplot(gs[row, :])
        else:
            ax = fig.add_subplot(gs[row, col_pos])

        ax.boxplot(data, vert=False)
        ax.axvline(lower, linestyle='--', label='IQR Lower')
        ax.axvline(upper, linestyle='--', label='IQR Upper')
        ax.set_title(col)
        ax.set_xlabel(col)

        if col_pos == 0:
            col_pos = 1
        else:
            col_pos = 0
            row += 1

    plt.tight_layout()
    plt.show()
```

### Outlier Handling

```python
df_outlier_treated = df_outlier.copy()

# CAPPING / WINSORIZATION: Years at Company & Promotions
capping_cols = ["Years at Company", "Monthly Income"]

print("\n[Capping / Winsorization]")
for col in capping_cols:
    if col in outlier_cols:
        lower = outlier_summary[col]["IQR_lower"]
        upper = outlier_summary[col]["IQR_upper"]

        before = ((df_outlier_treated[col] < lower) | 
                  (df_outlier_treated[col] > upper)).sum()

        df_outlier_treated[col] = df_outlier_treated[col].clip(lower, upper)

        after = ((df_outlier_treated[col] < lower) | 
                 (df_outlier_treated[col] > upper)).sum()

        print(f"- {col}: capped {before} -> {after} outliers")
```

*   **Capping/Winsorization** (cho **Years at Company** và **Promotions**): Đây là các biến dạng đếm. Việc cắt ngọn (Capping) giúp giữ lại thông tin về những nhân viên lâu năm/tài năng (thay vì xóa bỏ họ), đồng thời ngăn chặn các giá trị cực đại này gây nhiễu hoặc áp đảo trọng số của mô hình.

```python
outlier_cols = [
    col for col, info in outlier_summary.items()
    if (info["IQR_outliers"] > 0)
]

print(f"Columns with outliers: {len(outlier_cols)}")
print(outlier_cols)

if len(outlier_cols) == 0:
    print("No outliers detected!")
else:
    n_features = len(outlier_cols)
    n_rows = math.ceil(n_features / 2)

    fig = plt.figure(figsize=(14, 3.5 * n_rows))
    gs = GridSpec(n_rows, 2, figure=fig)

    row = col_pos = 0

    for i, col in enumerate(outlier_cols):
        data = df_outlier_treated[col].dropna()
        lower = outlier_summary[col]["IQR_lower"]
        upper = outlier_summary[col]["IQR_upper"]

        # Nếu là cột cuối và số lượng lẻ -> căn giữa
        if (i == n_features - 1) and (n_features % 2 == 1):
            ax = fig.add_subplot(gs[row, :])
        else:
            ax = fig.add_subplot(gs[row, col_pos])

        ax.boxplot(data, vert=False)
        ax.axvline(lower, linestyle='--', label='IQR Lower')
        ax.axvline(upper, linestyle='--', label='IQR Upper')
        ax.set_title(col)
        ax.set_xlabel(col)

        if col_pos == 0:
            col_pos = 1
        else:
            col_pos = 0
            row += 1

    plt.tight_layout()
    plt.show()
```

# Part 2:

## 2.1. Encoding Categorical Variables

```python
df_encoded = df_outlier_treated.copy()

# 1. Ordinal Encoding
ordinal_mappings = {
    'Work-Life Balance': ['Poor', 'Fair', 'Good', 'Excellent'],
    'Job Satisfaction': ['Low', 'Medium', 'High', 'Very High'],
    'Performance Rating': ['Below Average', 'Low', 'Average', 'High'],
    'Education Level': ['High School', 'Associate Degree', "Bachelor’s Degree", "Master’s Degree", 'PhD'],
    'Job Level': ['Entry', 'Mid', 'Senior'],
    'Company Reputation': ['Poor', 'Fair', 'Good', 'Excellent'],
    'Employee Recognition': ['Low', 'Medium', 'High', 'Very High']
}

for col, order in ordinal_mappings.items():
    df_encoded[col] = df_encoded[col].map({v: i for i, v in enumerate(order)})

# 2. Binary Encoding
binary_cols = [
    'Leadership Opportunities',
    'Innovation Opportunities',
    'Overtime',
    'Remote Work'
]

for col in binary_cols:
    df_encoded[col] = df_encoded[col].map({'No': 0, 'Yes': 1})

# 3. Target Encoding
df_encoded['Attrition'] = df_encoded['Attrition'].map({'Stayed': 0, 'Left': 1})

# 4. Nominal Encoding
nominal_cols = ['Gender', 'Job Role', 'Marital Status', 'Company Size']
df_onehot = pd.get_dummies(df_encoded[nominal_cols], drop_first=True)
df_encoded = df_encoded.drop(columns=nominal_cols)
df_encoded = pd.concat([df_encoded, df_onehot], axis=1)
```

```python
# ============================================================================
# KỊCH BẢN 1: Sau encoding, trước feature engineering
# ============================================================================
df_scenario1 = df_encoded.copy()

# Xác định các cột cần scaling
numeric_cols_scenario1 = df_scenario1.select_dtypes(include=[np.number]).columns.tolist()
numeric_cols_scenario1.remove('Attrition')

# Các cột có phân phối lệch phải -> log transform
log_cols_scenario1 = ['Monthly Income']

# Các cột cần StandardScaler
std_cols_scenario1 = [col for col in numeric_cols_scenario1 if col not in log_cols_scenario1]
scale_cols_scenario1 = log_cols_scenario1 + std_cols_scenario1

# Train-test-val split
X_scenario1 = df_scenario1.drop(columns=['Attrition'])
y_scenario1 = df_scenario1['Attrition']

X_train_s1, X_temp_s1, y_train_s1, y_temp_s1 = train_test_split(
    X_scenario1, y_scenario1, test_size=0.3, stratify=y_scenario1, random_state=1
)
X_test_s1, X_val_s1, y_test_s1, y_val_s1 = train_test_split(
    X_temp_s1, y_temp_s1, test_size=0.5, stratify=y_temp_s1, random_state=1
)

# Log transform
X_train_s1[log_cols_scenario1] = np.log1p(X_train_s1[log_cols_scenario1])
X_val_s1[log_cols_scenario1] = np.log1p(X_val_s1[log_cols_scenario1])
X_test_s1[log_cols_scenario1] = np.log1p(X_test_s1[log_cols_scenario1])

# Scaling
scaler_s1 = StandardScaler()
X_train_s1[scale_cols_scenario1] = scaler_s1.fit_transform(X_train_s1[scale_cols_scenario1])
X_val_s1[scale_cols_scenario1] = scaler_s1.transform(X_val_s1[scale_cols_scenario1])
X_test_s1[scale_cols_scenario1] = scaler_s1.transform(X_test_s1[scale_cols_scenario1])

# Save kịch bản 1
base_path_s1 = "../data/scenarios/scenario1/train_test"
os.makedirs(base_path_s1, exist_ok=True)

train_df_s1 = X_train_s1.copy()
train_df_s1["Attrition"] = y_train_s1.values
val_df_s1 = X_val_s1.copy()
val_df_s1["Attrition"] = y_val_s1.values
test_df_s1 = X_test_s1.copy()
test_df_s1["Attrition"] = y_test_s1.values

train_df_s1.to_csv(f"{base_path_s1}/train.csv", index=False)
val_df_s1.to_csv(f"{base_path_s1}/val.csv", index=False)
test_df_s1.to_csv(f"{base_path_s1}/test.csv", index=False)

print("Kịch bản 1 - Shape:", df_scenario1.shape)
print("Kịch bản 1 - Train shape:", X_train_s1.shape)
print("Kịch bản 1 - Test shape :", X_test_s1.shape)
print("Kịch bản 1 - Val shape :", X_val_s1.shape)
print(f"Đã lưu kịch bản 1 vào: {base_path_s1}")

# Kfold cho kịch bản 1
df_kfold_s1 = df_scenario1.copy()
df_kfold_s1[log_cols_scenario1] = np.log1p(df_kfold_s1[log_cols_scenario1])

X_kf_s1 = df_kfold_s1.drop(columns=["Attrition"])
y_kf_s1 = df_kfold_s1["Attrition"]

skf_s1 = StratifiedKFold(n_splits=5, shuffle=True, random_state=1)

base_path_kfold_s1 = "../data/scenarios/scenario1/kfold"
os.makedirs(base_path_kfold_s1, exist_ok=True)

for fold, (train_idx, val_idx) in enumerate(skf_s1.split(X_kf_s1, y_kf_s1), 1):
    X_train_kf_s1 = X_kf_s1.iloc[train_idx].copy()
    X_val_kf_s1 = X_kf_s1.iloc[val_idx].copy()
    y_train_kf_s1 = y_kf_s1.iloc[train_idx]
    y_val_kf_s1 = y_kf_s1.iloc[val_idx]
    
    # Fit scaler trên TRAIN của fold
    scaler_kf_s1 = StandardScaler()
    X_train_kf_s1[scale_cols_scenario1] = scaler_kf_s1.fit_transform(X_train_kf_s1[scale_cols_scenario1])
    X_val_kf_s1[scale_cols_scenario1] = scaler_kf_s1.transform(X_val_kf_s1[scale_cols_scenario1])
    
    # Gộp X + y
    train_fold_s1 = X_train_kf_s1.copy()
    train_fold_s1["Attrition"] = y_train_kf_s1.values
    val_fold_s1 = X_val_kf_s1.copy()
    val_fold_s1["Attrition"] = y_val_kf_s1.values
    
    # Save
    train_fold_s1.to_csv(f"{base_path_kfold_s1}/train_fold_{fold}.csv", index=False)
    val_fold_s1.to_csv(f"{base_path_kfold_s1}/val_fold_{fold}.csv", index=False)
    
    print(f"Fold {fold}: Train={len(train_idx)}, Val={len(val_idx)}")
```

## 2.2. Feature Engineering

### New features

```python
df_fe = df_encoded.copy()

# 1. Promotion speed
df_fe['Years_Per_Promotion'] = df_fe['Years at Company'] / (df_fe['Number of Promotions'] + 1)

# 2. Income relative to age
df_fe['Income_Per_Age'] = df_fe['Monthly Income'] / df_fe['Age']

# 3. Overall satisfaction score
df_fe['Total_Satisfaction'] = (df_fe['Job Satisfaction'] + df_fe['Work-Life Balance'])

# 4. Commute stress 
df_fe['Commute_Stress'] = df_fe['Distance from Home'] * (1 - df_fe['Remote Work'])

# 5. Tenure ratio
df_fe['Tenure_Ratio'] = df_fe['Years at Company'] / df_fe['Age']
```

Dưới đây là giải thích ý nghĩa cho từng đặc trưng vừa tạo:

1. Chỉ số dậm chân tại chỗ (**Years_Per_Promotion**)
*   **Công thức:** $\frac{\text{Years at Company}}{\text{Number of Promotions} + 1}$
*   Phản ánh tốc độ thăng tiến trung bình của nhân viên trong suốt thời gian làm việc tại công ty. Giá trị càng cao cho thấy nhân viên phải chờ đợi lâu để được thăng chức, từ đó dễ cảm nhận sự đình trệ trong sự nghiệp, giảm động lực làm việc và gia tăng khả năng nghỉ việc.

2. Áp lực tài chính theo độ tuổi (**Income_Per_Age**)
*   **Công thức:** $\frac{\text{Monthly Income}}{\text{Age}}$
*   Đánh giá mức thu nhập tương đối của nhân viên so với độ tuổi và giai đoạn cuộc sống. Cùng một mức lương có thể là phù hợp với người trẻ nhưng lại gây bất mãn với nhân viên lớn tuổi đang gánh vác nhiều trách nhiệm tài chính, từ đó ảnh hưởng đến ý định gắn bó lâu dài với công ty.

3. Tổng điểm hài lòng (**Total_Satisfaction**)
*   **Công thức:** **Job Satisfaction** + **Work-Life Balance**
*   Phản ánh mức độ hài lòng tổng thể của nhân viên, bao gồm cả yếu tố công việc và cân bằng cuộc sống. Giá trị thấp cho thấy sự bất mãn trên nhiều khía cạnh cùng lúc, thường liên quan đến trạng thái kiệt sức (burnout) hoặc mất động lực, là tín hiệu mạnh dẫn đến quyết định nghỉ việc.

4. Áp lực di chuyển thực tế (**Commute_Stress**)
*   **Công thức:** **Distance from Home $\times$ $(1 - \text{Remote Work})$**
*   Mô hình hóa áp lực đi lại thực tế bằng cách kết hợp khoảng cách địa lý và chế độ làm việc. Với nhân viên làm việc từ xa, áp lực đi lại gần như bằng không, trong khi với nhân viên làm tại văn phòng, quãng đường di chuyển dài có thể làm gia tăng căng thẳng và giảm mức độ hài lòng công việc.


5. Tỷ lệ cống hiến (**Tenure_Ratio**)
*   **Công thức:** $\frac{\text{Years at Company}}{\text{Age}}$
*   Đo lường mức độ gắn bó của nhân viên với công ty thông qua tỷ lệ thời gian làm việc so với tuổi đời. Giá trị cao thường cho thấy sự trung thành hoặc tâm lý ngại thay đổi, trong khi giá trị thấp có thể phản ánh xu hướng dịch chuyển nghề nghiệp, từ đó ảnh hưởng đến xác suất nghỉ việc.

```python
corr_matrix = df_fe.corr().abs()
target_corr = corr_matrix["Attrition"]
upper_triangle = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

threshold = 0.85
to_drop = set()

for col in upper_triangle.columns:
    high_corr = upper_triangle[col][upper_triangle[col] >= threshold]

    for row in high_corr.index:
        # So sánh correlation với target
        if target_corr[col] < target_corr[row]:
            to_drop.add(col)
        else:
            to_drop.add(row)

to_drop = list(to_drop)

print(f"Number of features to drop (|corr| >= {threshold}): {len(to_drop)}")
print(to_drop)
```
