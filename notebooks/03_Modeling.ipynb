### Chiến lược thực nghiệm (Experimentation Strategy)

**Two-Stage Efficient Approach:**

| Giai đoạn | Method | Data | Models | Mục đích |
|-----------|--------|------|--------|----------|
| **Stage 1: Screening** | Train-Test Split | train_test folder | All 11 models | Screen all models, select top-k finalists |
| **Stage 2: Final Evaluation** | 5-Fold CV | kfold folder | Top finalists only | Robust evaluation, choose champion |

**Selection Strategy:**
- **Linear Models** (4 models) → Top 2: Best performing
- **Tree Models** (3 models) → Top 2: Best performing  
- **Ensemble Models** (4 models) → Top 2: Best performing
- **Total Finalists:** 6 models per scenario

---

### Lập luận chọn Metric chính (Primary Metric Selection)

#### **Phân tích đặc điểm bài toán**

Business Context:**
```
Positive Class (Left): Nhân viên nghỉ việc
Negative Class (Stayed): Nhân viên ở lại
```

Cost Analysis:**

| Loại sai | Tên | Hậu quả | Mức độ nghiêm trọng |
|----------|-----|---------|---------------------|
| **False Negative (FN)** | Dự đoán "Stayed" nhưng thực tế "Left" | **BỎ SÓT** nhân viên có nguy cơ nghỉ → Mất nhân tài quan trọng | **RẤT NGHIÊM TRỌNG** |
| **False Positive (FP)** | Dự đoán "Left" nhưng thực tế "Stayed" | Chi phí can thiệp giữ chân không cần thiết | **Chấp nhận được** |

####  **Yêu cầu nghiệp vụ**
> **"Ưu tiên dự đoán đúng nghỉ việc (Left) hơn là dự đoán đúng ở lại (Stayed)"**

**→ Điều này có nghĩa là:**
- **Không bỏ sót** nhân viên có nguy cơ rời đi (Minimize False Negatives)
- **Phát hiện được càng nhiều người nghỉ việc càng tốt** (High Recall/Sensitivity)
-  Có thể chấp nhận một số False Positives (dự đoán nhầm người ở lại thành sẽ nghỉ)

---



##### ** F2-Score — Nhấn mạnh Recall nhưng vẫn cân nhặc Precision**
$$F_2 = \frac{(1 + 2^2) \times \text{Precision} \times \text{Recall}}{(2^2 \times \text{Precision}) + \text{Recall}} = \frac{5 \times \text{Precision} \times \text{Recall}}{(4 \times \text{Precision}) + \text{Recall}}$$

**Đặc điểm:**
- Cân bằng giữa Precision và Recall, nhưng **weight Recall gấp 2 lần Precision**
- Ưu tiên Recall (phát hiện người nghỉ việc)
- Vẫn kiểm soát Precision (tránh dự đoán sai quá nhiều)

````python
# Helper Functions

def load_train_test_data(scenario_path: str) -> Dict:
    """
    Load train và test data từ folder train_test
    Args:
        scenario_path: Đường dẫn đến scenario folder
    Returns:
        Dictionary chứa train và test DataFrames
    """
    folder = os.path.join(scenario_path, 'train_test')
    train_df = pd.read_csv(os.path.join(folder, 'train.csv'))
    test_df = pd.read_csv(os.path.join(folder, 'test.csv'))
    print(f"Loaded train-test data:")
    print(f"   Train: {train_df.shape} | Test: {test_df.shape}")
    return {'train': train_df, 'test': test_df}


def load_kfold_data(scenario_path: str) -> list:
    """
    Load toàn bộ các cặp train/val của k-fold đã chia sẵn
    Args:
        scenario_path: Đường dẫn đến scenario folder
    Returns:
        List các tuple (train_df, val_df) cho từng fold
    """
    folder = os.path.join(scenario_path, 'kfold')
    folds = []
    for i in range(1, 6):
        train_fold = pd.read_csv(os.path.join(folder, f'train_fold_{i}.csv'))
        val_fold = pd.read_csv(os.path.join(folder, f'val_fold_{i}.csv'))
        folds.append((train_fold, val_fold))
    print(f"Loaded {len(folds)} folds from kfold directory.")
    return folds


def prepare_data(df: pd.DataFrame, target_col: str = 'Attrition') -> Tuple:
    """
    Chuẩn bị data cho training (KHÔNG encode lại vì preprocessing đã làm rồi)
    Args:
        df: DataFrame
        target_col: Tên cột target
    Returns:
        X, y
    """
    y = df[target_col].values
    X = df.drop(columns=[target_col])
    return X, y


def cross_validate_model_with_ready_folds(model, folds: list, target_col: str = 'Attrition') -> dict:
    """
    Cross-validation sử dụng các fold đã chia sẵn (không dùng StratifiedKFold)
    Args:
        model: Model object
        folds: List các tuple (train_df, val_df)
        target_col: Tên cột target
    Returns:
        Dictionary chứa metrics với mean, std và chi tiết từng fold
    """
    scores = {
        'accuracy': [],
        'precision': [],
        'recall': [],
        'f1': [],
        'f2': [],
        'roc_auc': []
    }
    for train_df, val_df in folds:
        X_train, y_train = prepare_data(train_df, target_col)
        X_val, y_val = prepare_data(val_df, target_col)
        model.fit(X_train, y_train)
        y_pred = model.predict(X_val)
        y_pred_proba = model.predict_proba(X_val)[:, 1] if hasattr(model, 'predict_proba') else None
        scores['accuracy'].append(accuracy_score(y_val, y_pred))
        scores['precision'].append(precision_score(y_val, y_pred))
        scores['recall'].append(recall_score(y_val, y_pred))
        scores['f1'].append(f1_score(y_val, y_pred))
        scores['f2'].append(fbeta_score(y_val, y_pred, beta=2))
        if y_pred_proba is not None:
            scores['roc_auc'].append(roc_auc_score(y_val, y_pred_proba))
    results = {}
    for metric in scores.keys():
        results[f'{metric}_mean'] = np.mean(scores[metric])
        results[f'{metric}_std'] = np.std(scores[metric])
        results[f'{metric}_folds'] = scores[metric]
    return results


def select_top_models_by_group(results_df: pd.DataFrame, scenario: str) -> List[str]:
    """
    Chọn top models từ mỗi nhóm dựa trên F2-Score (PRIMARY METRIC)
    Strategy:
    - Linear Models (4 models) → Top 2: Best performing
    - Tree Models (3 models) → Top 3: Best performing
    - Ensemble Models (4 models) → Top 3: Best performing
    - Total Finalists: 8 models per scenario
    Args:
        results_df: DataFrame chứa kết quả screening
        scenario: scenario_key (ví dụ: 'scenario1')
    Returns:
        List các model keys được chọn
    """
    scenario_data = results_df[results_df['scenario_key'] == scenario].copy()
    finalists = []
    # Group 1: Linear (4 models)
    linear_keys = ['lr', 'ridge', 'sgd', 'lsvc']
    linear_models = scenario_data[scenario_data['model_key'].isin(linear_keys)]
    if len(linear_models) > 0:
        top_linear = linear_models.nlargest(2, 'f2')
        finalists.extend(top_linear['model_key'].tolist())
    # Group 2: Tree (3 models)
    tree_keys = ['dt', 'et', 'gb']
    tree_models = scenario_data[scenario_data['model_key'].isin(tree_keys)]
    if len(tree_models) > 0:
        top_tree = tree_models.nlargest(3, 'f2')
        finalists.extend(top_tree['model_key'].tolist())
    # Group 3: Ensemble (4 models)
    ensemble_keys = ['rf', 'xgb', 'lgbm', 'catboost']
    ensemble_models = scenario_data[scenario_data['model_key'].isin(ensemble_keys)]
    if len(ensemble_models) > 0:
        top_ensemble = ensemble_models.nlargest(3, 'f2')
        finalists.extend(top_ensemble['model_key'].tolist())
    return finalists


print("Helper functions defined!")
````

## 3. Stage 1: Model Screening

````python
# Stage 1: Model Screening
print("=" * 80)
print("STAGE 1: MODEL SCREENING")
print("=" * 80)
print("Objective: Screen all 6 models and select top finalists from each group")
print("Method: Train-Test Split (train_test folder)")
print("=" * 80)

results_stage1 = []

for scenario_key, scenario_info in CONFIG['scenarios'].items():
    print(f"\n{scenario_info['name']}")
    print(f"   {scenario_info['description']}")
    print("-" * 80)
    # Load 
    data = load_train_test_data(scenario_info['path'])
    
    X_train, y_train = prepare_data(data['train'])
    X_test, y_test = prepare_data(data['test'])
    
    print(f"\n   Features: {X_train.shape[1]}")
    print(f"   Train: {len(X_train)} | Test: {len(X_test)} samples")
    
    for model_key, model_info in CONFIG['models'].items():
        print(f"\n Screening {model_info['name']}...", end=" ")

        from sklearn.base import clone
        model = clone(model_info['model'])
        
        # Train
        start_time = time.time()
        model.fit(X_train, y_train)
        train_time = time.time() - start_time
        
        y_pred = model.predict(X_test)
        y_pred_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, 'predict_proba') else None
        
        # Metrics
        result = {
            'scenario': scenario_info['name'],
            'scenario_key': scenario_key,
            'model': model_info['name'],
            'model_key': model_key,
            'num_features': X_train.shape[1],
            'train_samples': len(X_train),
            'test_samples': len(X_test),
            'accuracy': accuracy_score(y_test, y_pred),
            'precision': precision_score(y_test, y_pred),
            'recall': recall_score(y_test, y_pred),
            'f1': f1_score(y_test, y_pred),
            'f2': fbeta_score(y_test, y_pred, beta=2),
            'roc_auc': roc_auc_score(y_test, y_pred_proba) if y_pred_proba is not None else None,
            'train_time': train_time
        }
        
        results_stage1.append(result)
        
        print(f"F2: {result['f2']:.4f} | Recall: {result['recall']:.4f} | Precision: {result['precision']:.4f} | Accuracy: {result['accuracy']:.4f} | F1: {result['f1']:.4f}")

# Convert to DataFrame
df_stage1 = pd.DataFrame(results_stage1)
print("\n\nStage 1 completed!")
print(f"   Total screenings: {len(results_stage1)}")
````

````python
# Chọn finalists cho từng scenario sau Stage 1
finalists_by_scenario = {}
print("\n[DEBUG] df_stage1['scenario_key'].unique():", df_stage1['scenario_key'].unique())
for scenario_key in CONFIG['scenarios']:
    print(f"\n[DEBUG] Đang xét scenario_key: {scenario_key}")
    scenario_data = df_stage1[df_stage1['scenario_key'] == scenario_key].copy()
    print(f"[DEBUG] Số dòng tìm được: {len(scenario_data)}")
    if len(scenario_data) > 0:
        print(scenario_data[['model_key','f2']])
    finalists = select_top_models_by_group(df_stage1, scenario_key)
    print(f"[DEBUG] Finalists: {finalists}")
    finalists_by_scenario[scenario_key] = finalists
print("\nFinalists by scenario:")
for k, v in finalists_by_scenario.items():
    print(f"  {k}: {v}")
````

````python
# Stage 2: Cross-Validation of Finalists
print("=" * 80)
print("STAGE 2: CROSS-VALIDATION OF FINALISTS")
print("=" * 80)
print("Objective: Robust evaluation of finalists to find champion model")
print("Method: 5-Fold Cross-Validation trên các fold đã chia sẵn (kfold folder)")
print("=" * 80)

results_stage2 = []

for scenario_key, scenario_info in CONFIG['scenarios'].items():
    print(f"\n{scenario_info['name']}")
    print(f"   {scenario_info['description']}")
    print("-" * 80)
    
    finalists = finalists_by_scenario[scenario_key]
    print(f"\n   Finalists: {len(finalists)} models")
    
    folds = load_kfold_data(scenario_info['path'])
    print(f"   Features: {folds[0][0].shape[1]}")
    print(f"   Total samples: {sum(len(train)+len(val) for train,val in folds)} (5-Fold CV)")
    
    for model_key in finalists:
        model_info = CONFIG['models'][model_key]
        print(f"\n   🔄 CV {model_info['name']}...", end=" ")
        
        from sklearn.base import clone
        model = clone(model_info['model'])

        cv_results = cross_validate_model_with_ready_folds(model, folds, target_col='Attrition')
        
        f2_mean = cv_results['f2_mean']
        f2_folds = cv_results['f2_folds']
        worst_degradation = max(f2_mean - f for f in f2_folds)
        cv_results['f2_worst_degradation'] = worst_degradation
        
        cv_results['scenario'] = scenario_info['name']
        cv_results['scenario_key'] = scenario_key
        cv_results['model'] = model_info['name']
        cv_results['model_key'] = model_key
        cv_results['num_features'] = folds[0][0].shape[1]
        
        results_stage2.append(cv_results)
        
        print(f"F2: {cv_results['f2_mean']:.4f} (±{cv_results['f2_std']:.4f}) | Worst degradation: {worst_degradation:.4f} | Recall: {cv_results['recall_mean']:.4f}")
        print("      Fold F2-scores:", ", ".join([f"{score:.4f}" for score in f2_folds]))

# Convert to DataFrame
df_stage2 = pd.DataFrame(results_stage2)
print("\n\nStage 2 completed!")
print(f"   Total CV evaluations: {len(results_stage2)}")

print("\n\n" + "=" * 100)
print("DETAILED FOLD-BY-FOLD RESULTS")
print("=" * 100)

for scenario_key, scenario_info in CONFIG['scenarios'].items():
    print(f"\n{scenario_info['name']}")
    print("-" * 100)
    
    scenario_results = [r for r in results_stage2 if r['scenario_key'] == scenario_key]
    
    for result in scenario_results:
        print(f"\n  {result['model']}")
        print(f"     Mean F2: {result['f2_mean']:.4f} ± {result['f2_std']:.4f}")
        print(f"     Worst degradation (F2): {result['f2_worst_degradation']:.4f}")
        print(f"     Mean Recall: {result['recall_mean']:.4f} ± {result['recall_std']:.4f}")
        print(f"     Mean Precision: {result['precision_mean']:.4f} ± {result['precision_std']:.4f}")
        print(f"\n     Fold-by-Fold F2 Scores:")
        print("       ", ", ".join([f"Fold {i+1}: {score:.4f}" for i, score in enumerate(result['f2_folds'])]))
        
        for i, fold_score in enumerate(result['f2_folds'], 1):
            deviation = fold_score - result['f2_mean']
            deviation_pct = (deviation / result['f2_mean']) * 100
            sign = "+" if deviation >= 0 else ""
            print(f"       Fold {i}: {fold_score:.4f}  ({sign}{deviation:+.4f} | {sign}{deviation_pct:+.2f}%)")
        
        print(f"\n     Summary Statistics:")
        print(f"       Best Fold:  {max(result['f2_folds']):.4f} (Fold {result['f2_folds'].index(max(result['f2_folds']))+1})")
        print(f"       Worst Fold: {min(result['f2_folds']):.4f} (Fold {result['f2_folds'].index(min(result['f2_folds']))+1})")
        print(f"       Range: {max(result['f2_folds']) - min(result['f2_folds']):.4f}")
        print(f"       Coefficient of Variation: {(result['f2_std'] / result['f2_mean'] * 100):.2f}%")
````

## 4. Stage 2: Cross-Validation of Finalists

Đánh giá robust với 5-Fold CV chỉ trên finalists để tìm champion model

````python
# Display Stage 2 Results & Champions
print("\n" + "=" * 140)
print("STAGE 2 CROSS-VALIDATION RESULTS")
print("=" * 140)

summary_cols = ['scenario', 'model', 'accuracy_mean', 'accuracy_std',
                'precision_mean', 'precision_std', 'recall_mean', 'recall_std',
                'f1_mean', 'f1_std', 'f2_mean', 'f2_std', 
                'roc_auc_mean', 'roc_auc_std', 'num_features']

df_summary_s2 = df_stage2[summary_cols].copy()

display_data = []
for _, row in df_summary_s2.iterrows():
    display_data.append({
        'Scenario': row['scenario'],
        'Model': row['model'],
        'Accuracy': f"{row['accuracy_mean']:.4f} ± {row['accuracy_std']:.4f}",
        'Precision': f"{row['precision_mean']:.4f} ± {row['precision_std']:.4f}",
        'Recall': f"{row['recall_mean']:.4f} ± {row['recall_std']:.4f}",
        'F1': f"{row['f1_mean']:.4f} ± {row['f1_std']:.4f}",
        'F2 ⭐': f"{row['f2_mean']:.4f} ± {row['f2_std']:.4f}",
        'ROC-AUC': f"{row['roc_auc_mean']:.4f} ± {row['roc_auc_std']:.4f}",
        'Features': int(row['num_features'])
    })

df_display_s2 = pd.DataFrame(display_data)
print(df_display_s2.to_string(index=False))

print("\n\n" + "=" * 80)
print("CHAMPION MODEL PER SCENARIO")
print("=" * 80)

champions = {}

for scenario in df_stage2['scenario'].unique():
    scenario_data = df_stage2[df_stage2['scenario'] == scenario]
    
    # Best by F2 mean (PRIMARY METRIC)
    best_idx = scenario_data['f2_mean'].idxmax()
    champion = scenario_data.loc[best_idx]
    
    scenario_key = champion['scenario_key']
    champions[scenario_key] = champion['model_key']
    
    print(f"\n{scenario}:")
    print(f"  🥇 CHAMPION: {champion['model']}")
    print(f"     F2 (PRIMARY): {champion['f2_mean']:.4f} ± {champion['f2_std']:.4f}")
    print(f"     Recall: {champion['recall_mean']:.4f} ± {champion['recall_std']:.4f}")
    print(f"     Precision: {champion['precision_mean']:.4f} ± {champion['precision_std']:.4f}")
    print(f"     F1: {champion['f1_mean']:.4f} ± {champion['f1_std']:.4f}")
    print(f"     Accuracy: {champion['accuracy_mean']:.4f} ± {champion['accuracy_std']:.4f}")
    print(f"     ROC-AUC: {champion['roc_auc_mean']:.4f} ± {champion['roc_auc_std']:.4f}")
    print(f"     Stability: {'Excellent' if champion['f2_std'] < 0.01 else 'Good' if champion['f2_std'] < 0.02 else 'Fair'}")
````

````python
from sklearn.metrics import fbeta_score, recall_score, make_scorer, precision_score, accuracy_score, roc_auc_score, f1_score
from sklearn.model_selection import GridSearchCV
from sklearn.base import clone
import numpy as np
import pandas as pd

param_grids = {
    'lgbm': {
        'n_estimators': [200, 500],
        'learning_rate': [0.05, 0.1],
        'num_leaves': [31, 63],
        'max_depth': [-1, 7],
        'min_child_samples': [20, 50]
    },
    'gb': {
        'n_estimators': [200, 300],
        'learning_rate': [0.05, 0.1],
        'max_depth': [3, 5],
        'subsample': [0.8, 1.0]
    },
    'rf': {
        'n_estimators': [100, 200],
        'max_depth': [None, 10, 20],
        'min_samples_split': [2, 5]
    },
    'lr': {
        'C': [0.1, 1.0, 10.0],
        'solver': ['lbfgs']
    }
}

f2_scorer = make_scorer(fbeta_score, beta=2)

nested_cv_results = []

for scenario_key, scenario_info in CONFIG['scenarios'].items():
    print(f"\n===== {scenario_info['name']} =====")
    
    if 'champions' not in locals():
        print("[ERROR] 'champions' dictionary not found. Run previous cells to select champion models.")
        break
    model_key = champions[scenario_key]
    model_info = CONFIG['models'][model_key]
    print(f"Champion model: {model_info['name']} ({model_key})")
    
    param_grid = param_grids.get(model_key, None)
    if param_grid is None:
        print(f"[SKIP] No param grid for {model_key}. Skipping grid search.")
        continue
    
    folds = load_kfold_data(scenario_info['path'])
    
    metrics_before = {'f2': [], 'recall': [], 'precision': [], 'accuracy': [], 'f1': [], 'roc_auc': []}
    metrics_after = {'f2': [], 'recall': [], 'precision': [], 'accuracy': [], 'f1': [], 'roc_auc': []}
    
    for fold_idx, (train_df, val_df) in enumerate(folds):
        print(f"\n--- Fold {fold_idx + 1} ---")
        X_train, y_train = prepare_data(train_df)
        X_val, y_val = prepare_data(val_df)
        
        base_model = clone(model_info['model'])
        base_model.fit(X_train, y_train)
        y_pred_base = base_model.predict(X_val)
        y_pred_proba_base = base_model.predict_proba(X_val)[:, 1] if hasattr(base_model, 'predict_proba') else None
        metrics_before['f2'].append(fbeta_score(y_val, y_pred_base, beta=2))
        metrics_before['recall'].append(recall_score(y_val, y_pred_base))
        metrics_before['precision'].append(precision_score(y_val, y_pred_base))
        metrics_before['accuracy'].append(accuracy_score(y_val, y_pred_base))
        metrics_before['f1'].append(f1_score(y_val, y_pred_base))
        if y_pred_proba_base is not None:
            metrics_before['roc_auc'].append(roc_auc_score(y_val, y_pred_proba_base))
        else:
            metrics_before['roc_auc'].append(np.nan)
        
        grid = GridSearchCV(
            estimator=clone(model_info['model']),
            param_grid=param_grid,
            scoring=f2_scorer,
            cv=3,
            n_jobs=-1,
            verbose=0
        )
        grid.fit(X_train, y_train)
        best_model = grid.best_estimator_
        y_pred_tuned = best_model.predict(X_val)
        y_pred_proba_tuned = best_model.predict_proba(X_val)[:, 1] if hasattr(best_model, 'predict_proba') else None
        metrics_after['f2'].append(fbeta_score(y_val, y_pred_tuned, beta=2))
        metrics_after['recall'].append(recall_score(y_val, y_pred_tuned))
        metrics_after['precision'].append(precision_score(y_val, y_pred_tuned))
        metrics_after['accuracy'].append(accuracy_score(y_val, y_pred_tuned))
        metrics_after['f1'].append(f1_score(y_val, y_pred_tuned))
        if y_pred_proba_tuned is not None:
            metrics_after['roc_auc'].append(roc_auc_score(y_val, y_pred_proba_tuned))
        else:
            metrics_after['roc_auc'].append(np.nan)
        
        print(f"F2: {metrics_before['f2'][-1]:.4f} → {metrics_after['f2'][-1]:.4f} | Recall: {metrics_before['recall'][-1]:.4f} → {metrics_after['recall'][-1]:.4f}")
    
    def summarize(metrics):
        return {
            'f2_mean': np.mean(metrics['f2']),
            'f2_std': np.std(metrics['f2']),
            'f2_worst_degradation': max(np.mean(metrics['f2']) - np.array(metrics['f2'])),
            'recall_mean': np.mean(metrics['recall']),
            'recall_std': np.std(metrics['recall']),
            'precision_mean': np.mean(metrics['precision']),
            'precision_std': np.std(metrics['precision']),
            'accuracy_mean': np.mean(metrics['accuracy']),
            'accuracy_std': np.std(metrics['accuracy']),
            'f1_mean': np.mean(metrics['f1']),
            'f1_std': np.std(metrics['f1']),
            'roc_auc_mean': np.nanmean(metrics['roc_auc']),
            'roc_auc_std': np.nanstd(metrics['roc_auc']),
            'f2_folds': metrics['f2'],
            'recall_folds': metrics['recall']
        }
    
    summary_before = summarize(metrics_before)
    summary_after = summarize(metrics_after)
    
    print("\nSummary BEFORE tuning:")
    print(f"F2: {summary_before['f2_mean']:.4f} ± {summary_before['f2_std']:.4f} | Worst degradation: {summary_before['f2_worst_degradation']:.4f} | Recall: {summary_before['recall_mean']:.4f}")
    print("Summary AFTER tuning:")
    print(f"F2: {summary_after['f2_mean']:.4f} ± {summary_after['f2_std']:.4f} | Worst degradation: {summary_after['f2_worst_degradation']:.4f} | Recall: {summary_after['recall_mean']:.4f}")
    
    nested_cv_results.append({
        'scenario': scenario_info['name'],
        'scenario_key': scenario_key,
        'model': model_info['name'],
        'model_key': model_key,
        'num_features': folds[0][0].shape[1],
        'before': summary_before,
        'after': summary_after
    })

display_rows = []
for res in nested_cv_results:
    display_rows.append({
        'Scenario': res['scenario'],
        'Model': res['model'],
        'F2 (Before)': f"{res['before']['f2_mean']:.4f} ± {res['before']['f2_std']:.4f}",
        'F2 (After)': f"{res['after']['f2_mean']:.4f} ± {res['after']['f2_std']:.4f}",
        'Recall (Before)': f"{res['before']['recall_mean']:.4f}",
        'Recall (After)': f"{res['after']['recall_mean']:.4f}",
        'Worst Degradation (Before)': f"{res['before']['f2_worst_degradation']:.4f}",
        'Worst Degradation (After)': f"{res['after']['f2_worst_degradation']:.4f}",
        'Features': res['num_features']
    })
df_nested_cv = pd.DataFrame(display_rows)
print("\n=== Nested CV Results (Tuned vs. Untuned) ===")
print(df_nested_cv.to_string(index=False))
````
